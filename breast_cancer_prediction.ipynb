```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:,1]= labelencoder_Y.fit_transform(df.iloc[:,1].values)
print(labelencoder_Y.fit_transform(df.iloc[:,1].values))
```

```python
X = df.iloc[:, 2:31].values 
Y = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
print(X_train)
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
NC=KNeighborsClassifier(n_neighbors=4)
NC.fit(X_train,Y_train)
```

```python
# predict training
```

```python
from sklearn.metrics import accuracy_score
predict_training_data=NC.predict(X_train)
accuracy_train=accuracy_score(Y_train,predict_training_data)
print("ACCURACY ON TRAINING DATA:",accuracy_train)
```

```python
#predict testing
```

```python
predict_testing_data=NC.predict(X_test)
accuracy_test=accuracy_score(Y_test,predict_testing_data)
print("ACCURACY ON Testing DATA:",accuracy_test)
```

```python
#SVM
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
predict_training=clf.predict(X_train)
accuracy_train=accuracy_score(Y_train,predict_training)
print("ACCURACY ON TRAINING DATA-SVM:",accuracy_train)
```

```python
# predict testing
```

```python
predict_test=clf.predict(X_test)
accuracy_test=accuracy_score(Y_test,predict_test)
print("ACCURACY ON TESTING DATA-SVM:",accuracy_test)
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

```python
#predict training
```

```python
from sklearn.metrics import accuracy_score
predict_training=log.predict(X_train)
accuracy_train=accuracy_score(Y_train,predict_training)
print("ACCURACY ON TRAINING DATA-LOGISTIC REGRESSION:",accuracy_train)
```

```python
#predict testing
```

```python
predict_test=log.predict(X_test)
accuracy_test=accuracy_score(Y_test,predict_test)
print("ACCURACY ON TESTING DATA-LOGISTIC REGRESSION:",accuracy_test)
```

```python
#RANDOM FOREST
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
forest.fit(X_train, Y_train)
  
```

```python
#Random forest
```

```python
from sklearn.metrics import accuracy_score
predict_training=forest.predict(X_train)
accuracy_train=accuracy_score(Y_train,predict_training)
print("ACCURACY ON TRAINING DATA-RANDOM FOREST CLASSIFIER :",accuracy_train)
```

```python
#predict testing:
```

```python
predict_test=forest.predict(X_test)
accuracy_test=accuracy_score(Y_test,predict_test)
print("ACCURACY ON TESTING DATA- RANDOM FOREST CLASSIFIER:",accuracy_test)
```
